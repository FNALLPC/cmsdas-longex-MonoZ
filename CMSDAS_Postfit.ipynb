# MonoZ Analysis - Prefit Vs Postit
Plotter for Higgs Combine FitDiagnostics and for MonoZ datacards
## Import Pre-requisites

```python
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
import uproot
import matplotlib._color_data as mcd
```

## Extract histograms 
We'll extract all histograms from the diagnostic file here and store them in a dictionary to be used down the line

```python
_verbose = True
#open the file and grab the histograms
fn = '/eos/user/r/rhashmi/work/CMSDAS2024/Combine/CMSSW_11_3_4/src/fitDiagnosticsMonoZ.root'
f = uproot.open(fn)
if (_verbose):
    print("\n-----------\nUprooted file\n------------\n")

# Function to recursively extract histograms from directories
def extract_histograms(directory, histograms):
    for key, classname in directory.classnames().items():
        if classname.startswith("TDirectory"):
            subdir = directory[key]
            extract_histograms(subdir, histograms)
        elif classname in ["TH1F", "TH1D", "TGraphAsymmErrors"]:
            try:
                obj = directory[key]
                histograms[key] = obj
                if (_verbose): 
                    print(f"Added {classname}: {key}")
            except Exception as e:
                print(f"Skipped {key} due to error: {e}")

# Dictionary to store histograms
histograms = {}

# Extract histograms from the root file
extract_histograms(f, histograms)

if (_verbose):
    print("\n----------\nConverted into histograms\n----------\n")

    # Histogram data
    for name, roothist in histograms.items():
        try:
            values = roothist.values()
            edges = roothist.axis().edges()
            print(f"Histogram: {name}")
            print(f"Values: {values}")
            print(f"Edges: {edges}")
        except Exception as e:
            print(f"Error processing {name}: {e}")
```

## Plotting distributions
### Channel 4L - Prefit
#### Extract data from relevent histograms

```python
_verbose = True
for name, roothist in histograms.items():
    name = name.replace(";1", "")
    era = "2016"
    channel = "cat4L"
    #channel = 'cat3L'
    fit_type = "prefit"
    values = roothist.values()
    
    # ren ame freq to bin_cont
    
    if  "shapes_"+fit_type+"/"+channel+"2016/ZZ" in name:
            ZZ_freq = np.abs(values)
            
    if  "shapes_"+fit_type+"/"+channel+"2016/WZ" in name:
            WZ_freq = np.abs(values)
            
    if  "shapes_"+fit_type+"/"+channel+"2016/WW" in name:
            WW_freq = np.abs(values)
            
    if  "shapes_"+fit_type+"/"+channel+"2016/TOP" in name:
            TOP_freq = np.abs(values)
            
    if  "shapes_"+fit_type+"/"+channel+"2016/DY" in name:
            DY_freq = np.abs(values)
            
    if  "shapes_"+fit_type+"/"+channel+"2016/VVV" in name:
            VVV_freq = np.abs(values)
            
    if  "shapes_"+fit_type+"/"+channel+"2016/total_background" in name:
            bg_freq = np.abs(values)
            MCsyslow = roothist.errors()
            MCsysup = roothist.errors()

    if  "shapes_"+fit_type+"/"+channel+"2016/data" in name:
            # TGraphAsymmErrors values has two arrays [[x pos], [y pos]]
            x_values, data_set = values
            #Calculate Poisson errors
            #alpha = 1. - 0.6827
            #L = np.where(data_set == 0, 0, stats.gamma.ppf(alpha / 2, data_set, scale=1.))
            #U = stats.gamma.ppf(1 - alpha / 2, data_set + 1, scale=1.)
            Ylow = roothist.errors('low')[1]
            #Ylow = data_set - L
            #Yhigh = U - data_set
            Yhigh = roothist.errors('high')[1]
            
# No WW case so no entries. Setting to 0
WW_freq = ZZ_freq * 0

if (_verbose):
    print("ZZ bg: ", ZZ_freq)
    print("WZ bg: ", WZ_freq)
    print("WW bg: ", WW_freq)
    print("TOP bg: ", TOP_freq)
    print("DY bg: ", DY_freq)
    print("VVV bg: ", VVV_freq)
    print("\n")
    print("Tot bg: ", bg_freq)
    print("Tot bg Low: ", MCsyslow)
    print("Tot bg Up: ", MCsysup)
    print("\n")
    print("data: ", data_set)
    print("data Low ", Ylow)
    print("data Up", Yhigh)
```

#### Trimming arrays before plotting
Check relevant shapes ROOT file for number of bins and bin edges

```python
ZZ_freq = ZZ_freq[:11]
WZ_freq = WZ_freq[:11]
WW_freq = WW_freq[:11]
DY_freq = DY_freq[:11]
TOP_freq = TOP_freq[:11]
VVV_freq = VVV_freq[:11]

bg_freq = bg_freq[:11]
MCsyslow = MCsyslow[:11]
MCsysup = MCsysup[:11]

data_set = data_set[:11]
Yhigh = Yhigh[:11]
Ylow = Ylow[:11]
```

#### Set-up the bins and make the plots!

```python
bins = np.array([100,125,150,175,200,250,300,350,400,500,600,1000])
binc = np.array([ 0.5*(bins[i]+bins[i+1])for i in range(bins.size - 1)])
xerr = np.diff(bins)*0.5

#Set up variables for the stacked histogram
#bg_freq = VVV_freq + ZZ_freq + WW_freq + WZ_freq + TOP_freq + DY_freq
freqs = [VVV_freq, ZZ_freq, WW_freq, WZ_freq, TOP_freq, DY_freq]
binsc = [binc, binc, binc, binc, binc, binc]
proc_name = ["VVV", "ZZ", "WW", "WZ", "TOP", "DY"]
colors = ["maroon", "gold", 
          mcd.XKCD_COLORS["xkcd:green"].upper(), 
          mcd.XKCD_COLORS["xkcd:azure"].upper(), 
          mcd.XKCD_COLORS["xkcd:darkblue"].upper(), 
          mcd.XKCD_COLORS["xkcd:pink"].upper()]

#Set up variables for the stacked histogram
plt.figure(figsize=(8,10))
plt.gcf().subplots_adjust(bottom=0.15, left=0.17)
ax1 = plt.subplot2grid((4,1), (0,0),rowspan=2)

#the histograms (SM,Data)
#print(bins,'\n',freqs,'\n',colors,'\n',proc_name)
ax1.hist(binsc,bins=bins,weights=freqs,stacked=True, color=colors, label=proc_name)
ax1.errorbar(binc,data_set, xerr=xerr, yerr = [Ylow,Yhigh],color="black", fmt='o')
ax1.bar(binc,MCsyslow+MCsysup,bottom=bg_freq-MCsyslow,color='gray', hatch='',width=xerr*2,alpha=0.3)

#Labels and limits
ax1.set_yscale("log")
ax1.set_ylim(0.001,5000)
ax1.legend(loc="upper right", fontsize=14)
ax1.set_ylabel(r"$events$")
cms = ax1.text(
        0.5, 5500., u"CMS $\it{Preliminary}$",
        fontsize=16, fontweight='bold',
    )
lumi = ax1.text(
        800, 5500., r"%.1f fb$^{-1}$ (13 TeV)" % 35.9,
        fontsize=14, 
    )


#Adding in the ratio plot
ax2 = plt.subplot2grid((4,1), (2,0),sharex=ax1)
plt.setp(ax1.get_xticklabels(), visible=False)
#ax2.errorbar(binc,data_set/bg_freq, xerr=xerr,yerr=[data_set/bg_freq - (data_set-Ylow)/bg_freq,(data_set+Yhigh)/bg_freq - data_set/bg_freq], color="black", fmt='o')
ax2.errorbar(binc,data_set/bg_freq, xerr=xerr,yerr=[Ylow/bg_freq,Yhigh/bg_freq], color="black", fmt='o')

ax2.bar(binc,(MCsyslow+MCsysup)/bg_freq,bottom=1-MCsyslow/bg_freq,color='gray',width=xerr*2,alpha=0.3,hatch='')
ax2.axhline(1, ls="--")
ax2.set_ylim(0,4)
ax2.set_xlabel("$p_{T}^{miss}$ (GeV)",ha="right")
#ax2.set_xlabel("$M_{T}$ (GeV)",ha="right")
ax2.set_ylabel(r"Data/$\Sigma$MC")
```
